#Q6 )Write a code that generates the squares of numbers from 1 to n using a generator

```python
#Ans)
def squares(n):
  """
  This generator function yields the squares of numbers from 1 to n.
  """
  for i in range(1, n + 1):
    yield i ** 2

# Generate squares from 1 to 5
for square in squares(5):
  print(square)
```

#Q7 )Write a code that generates palindromic numbers up to n using a generator

```python
#Ans)

def palindrome_generator(n):
  """
  This generator function yields palindromic numbers up to n.
  """
  for num in range(1, n + 1):
    # Check if the number is the same forwards and backwards
    if str(num) == str(num)[::-1]:
      yield num

# Generate palindromic numbers up to 50
for palindrome in palindrome_generator(50):
  print(palindrome)
```

#Q8 )Write a code that generates even numbers from 2 to n using a generator

```python
#Ans)
def even_generator(n):
  """
  This generator function yields even numbers from 2 to n.
  """
  i = 2
  while i <= n:
    yield i
    i += 2

# Generate even numbers from 2 to 10
for even_number in even_generator(10):
  print(even_number)
```

#Q10 )Write a code that generates prime numbers up to n using a generator

```python
#Ans)
def prime_generator(n):
    """
    Generate prime numbers up to n.

    Args:
        n: The upper limit for the generated prime numbers.

    Yields:
        Prime numbers up to n.
    """
    # Start with first prime number.
    number = 2
    # Iterate until we reach the upper limit.
    while number <= n:
        # Check if the number is prime.
        is_prime = True
        for i in range(2, int(number**0.5) + 1):
            if number % i == 0:
                is_prime = False
                break
        # If the number is prime, yield it.
        if is_prime:
            yield number
        # Move on to the next number.
        number+=1
# Generate prime numbers up to 100.
for prime in prime_generator(51):
    print(prime)
```

#Q21 )Write a code that filters out prime numbers from a given list

```python
def is_prime(number):
  """
  This function checks if a number is prime.
  Args:
    number: The number to be checked.
  Returns:
    True if the number is prime, False otherwise.
  """
  if number <= 1:
    return False
  for divisor in range(2, int(number)):
    if number % divisor == 0:
      return False
  return True
def filter_primes(numbers):
  """
  This function filters out prime numbers from a given list.
  Args:
    numbers: The list of numbers.
  Returns:
    A new list containing only the prime numbers from the original list.
  """
  prime_numbers = []
  for number in numbers:
    if is_prime(number):
      prime_numbers.append(number)
  return prime_numbers
numbers = list(range(1,18))
prime_numbers = filter_primes(numbers)
print(prime_numbers)
```

#Q27 )What is encapsulation in OOP?

Encapsulation is a fundamental concept in object-oriented programming (OOP) that revolves around bundling data and methods together into a single unit called an object. It serves to protect the data from external interference and ensures that the object's internal state remains consistent.

Key Aspects of Encapsulation:

    1) Data Hiding: Encapsulation hides the internal details of an object from the outside world. This is achieved by declaring the data members of a class as private or protected. Private members are only accessible within the class, while protected members are accessible within the class and its subclasses.
    2) Information Hiding: Encapsulation conceals the implementation details of an object's methods. The user only interacts with the object through its public methods and does not need to know how these methods are implemented.
    3) Abstraction: Encapsulation abstracts the underlying complexity of an object by providing a simple interface for interacting with it. This simplifies the use of the object and makes it easier for developers to understand and maintain the code.
Benefits of Encapsulation:

    1) Improved Security: Encapsulation protects sensitive data from unauthorized access.
    2) Enhanced Modularity: Encapsulation allows for the creation of modular and reusable code components.
    3) Increased Code Maintainability: Encapsulation makes it easier to maintain and update code by isolating changes within the object.
    4) Reduced Coupling: Encapsulation reduces the interdependence between different parts of a program, making it more flexible and adaptable.

#Q31 )Explain method overriding in Python.

Method overriding in Python is a concept in object-oriented programming (OOP) that allows a child (subclass) to provide a specific implementation of a method that is already defined in its parent (superclass).


---

ðŸ§© Definition

When a subclass defines a method with the same name, parameters, and return type as a method in its parent class, the subclassâ€™s version overrides the parentâ€™s version.


---

âš™ï¸ How It Works

The base class (parent) defines a method.

The derived class (child) defines the same method name again â€” this replaces (overrides) the parent version.

When the method is called on a child class object, the childâ€™s version runs instead of the parentâ€™s.

```python
class Animal:
    def sound(self):
        print("Animal sound")
class Dog(Animal):
    def sound(self):        #method overriding
        print("Dog barks")
#create object
a=Animal()
d=Dog()
a.sound()
d.sound()
```

Here, the sound() method in Dog overrides the sound() method in Animal.

```python
#Sometimes, you might still want to use the parentâ€™s version inside the overridden method.

class Dog(Animal):
    def sound(self):
        super().sound()   # Call parent version
        print("Dog barks")
```

So, Overriding works only when inheritance is involved.Here the method name and parameters must be exactly the same. You can use super() to access the parent method.It helps to achieve polymorphism â€” the same method name behaves differently in different classes.

#Q37 )What is abstraction in Python? How is it implemented?

Abstraction in Python is a concept in object-oriented programming (OOP) that means hiding complex implementation details and showing only the essential features of an object.

It helps make programs simpler to understand and use by exposing only whatâ€™s necessary to the user.

ðŸ§  Example in Real Life:

When you drive a car, you use the steering wheel, pedals, and buttons â€” you donâ€™t need to know how the engine or transmission works internally.
Similarly, abstraction hides the â€œinternal mechanismâ€ of the code.
Abstraction is mainly implemented using the abc (Abstract Base Class) module.

This module allows us to:

1)Create abstract classes (classes that cannot be instantiated directly)

2)Define abstract methods (methods that must be implemented in child classes)

```python
from abc import ABC, abstractmethod

# Abstract class
class Animal(ABC):

    @abstractmethod
    def make_sound(self):
        pass  # Abstract method (no implementation)

# Concrete subclass
class Dog(Animal):
    def make_sound(self):
        return "Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

# Usage
dog = Dog()
cat = Cat()

print(dog.make_sound())  # Output: Woof!
print(cat.make_sound())  # Output: Meow!
```

#Q39 )How are abstract methods different from regular methods in Python?

Abstract method are different from regular methods in python as follows:

1)Declaration:
Abstract methods are declared within abstract classes using the @abstractmethod decorator from the abc module. Regular methods are declared within classes without any special decorator.

2)Implementation:
Abstract methods do not have any implementation in the abstract class where they are defined. Their purpose is to provide a method signature or interface that must be implemented by subclasses. Regular methods have concrete implementations within the class where they are defined. They perform specific actions or operations when called.

3)Purpose:
Abstract methods serve as placeholders for methods that must be implemented by subclasses. They define a contract that subclasses must adhere to. Regular methods provide functionality or behavior associated with the class where they are defined. They are directly callable and can be used to perform actions or manipulate data.

#Q40 )How can you achieve abstraction using interfaces in Python?

Abstraction means hiding implementation details and showing only essential features.
It allows you to focus on what an object does, not how it does it.
Python doesnâ€™t have a keyword like interface (as in Java),
but you can achieve the same behavior using abstract base classes (ABC).

An abstract class:

*)Can have abstract methods (methods with no implementation).

*)Can have concrete methods (with implementation).

*)Cannot be instantiated directly.

*)Must be subclassed by a concrete class that implements all abstract methods.

E.g.

```python
from abc import ABC, abstractmethod

# Define an abstract class (interface)
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass  # Abstract method - no implementation

    @abstractmethod
    def perimeter(self):
        pass

# Concrete class implementing the interface
class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        return self.length * self.breadth

    def perimeter(self):
        return 2 * (self.length + self.breadth)

# Using the interface
rect = Rectangle(5, 3)
print("Area:", rect.area())
print("Perimeter:", rect.perimeter())
```

Shape is an abstract base class â€” it defines the interface (area and perimeter) that all shapes must implement.

Rectangle provides concrete implementations of these methods.

We cannot create an object of Shape directly:

shape = Shape()  # âŒ This will raise an error

#Q41 )Can you provide an example of how abstraction can be utilized to create common interface for a group of related classes in Python?

```python
from abc import ABC, abstractmethod

# Define the common interface for animals
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

# Define concrete animal classes that implement the common interface
class Dog(Animal):
    def make_sound(self):
        return "Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

class Cow(Animal):
    def make_sound(self):
        return "Moo!"

# Usage example
def animal_sound(animal):
    if isinstance(animal, Animal):
        print(f"The {animal.__class__.__name__} says: {animal.make_sound()}")
    else:
        print("Not a valid animal!")

# Create instances of different animal classes
dog = Dog()
cat = Cat()
cow = Cow()

# Produce sounds of different animals
animal_sound(dog)  # Outputs: The Dog says: Woof!
animal_sound(cat)  # Outputs: The Cat says: Meow!
animal_sound(cow)  # Outputs: The Cow says: Moo!
```

#Q42 )How does Python achieve polymorphism through method overriding?

Python achieves polymorphism through method overriding by letting subclasses redefine parent class methods, so the same method call behaves differently based on the objectâ€™s actual type at runtime.
For e.g.

```python
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

class Cat(Animal):
    def speak(self):
        return "Cat meows"

# Demonstrating polymorphism
animals = [Dog(), Cat(), Animal()]

for a in animals:
    print(a.speak())
```

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand           # Public attribute
        self._model = model          # Protected attribute
        self.__mileage = 0           # Private attribute

    def drive(self):
        print("Driving...")

    def __update_mileage(self, miles):
        self.__mileage += miles

    def get_mileage(self):
        return self.__mileage

# Creating a car object
my_car = Car("Toyota", "Camry")

# Accessing protected attribute (convention, not enforced)
print(my_car._model)  # Output: Camry

# Accessing private attribute (name mangling)
# This is generally discouraged and not recommended
print(my_car._Car__mileage)  # Output: 0

# Calling public method
my_car.drive()  # Output: Driving...

# Calling private method (name mangling)
# Again, discouraged and not recommended
my_car._Car__update_mileage(100)
print(my_car.get_mileage())  # Output: 100
```

#Q56 )Create a decorator that preserves the metadata of the original function.

```python

import functools

def decorator(func):
    @functools.wraps(func)  # Preserve metadata of the original function
    def wrapper(*args, **kwargs):
        print("Decorator is executing")
        result = func(*args, **kwargs)  # Call the original function
        print("Decorator execution is completed")
        return result
    return wrapper

@decorator
def example_function(arg1, arg2):
    """
    This is an example function.
    """
    print(f"Arguments passed to the function: {arg1}, {arg2}")

# Print metadata of the decorated function
print("Function name:", example_function.__name__)
print("Function docstring:", example_function.__doc__)
```
